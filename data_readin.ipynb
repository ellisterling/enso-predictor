```python
import pandas as pd
import xarray as xr
```

```python
df_sst = xr.open_dataset("data/sst.mon.mean.nc")
df_sst
```

```python
df_sst.sst.sel(time = "2017-03-01").plot(vmin = 0)
```

```python
eq_pac = df_sst.sst.sel(lon = slice(100,300), lat = slice(50, -40))
eq_pac.sel(time = "2009-02-01").plot(vmin = 15)
```

```python
df_sst.sst.mean(dim = ["lat", "lon"]).plot()
```
